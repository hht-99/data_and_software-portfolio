```python
# Import script
import dogs_core

# Set the sdf it boots (same folder so just string name is okay, different folder = full path)
sdf_path = r"Enamine_Essential_Fragment_Library_320cmpds_20231002.sdf"
mols, smiles = dogs_core.load_fragments_from_sdf(sdf_path)
```

```python
# To draw outputs need to load dependencies 
from rdkit import Chem
from rdkit.Chem import AllChem, Draw

from rdkit import RDLogger
RDLogger.DisableLog('rdApp.*')

# Draw command
Draw.MolsToGridImage(mols[:5], molsPerRow=5, subImgSize=(200, 200))
```
