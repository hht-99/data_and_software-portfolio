```python
# Core RDKit and system tools
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
from rdkit.Chem.Draw import IPythonConsole

# Disable verbose RDKit logging
from rdkit import RDLogger
RDLogger.DisableLog('rdApp.*')
```

```python
# Set the full path to your fragment library
sdf_path = r"C:\Users\cm18hht\dogs_rebuild\Enamine_Essential_Fragment_Library_320cmpds_20231002.sdf"

# Function to load SDF and convert to list of molecules and SMILES
def load_fragments_from_sdf(sdf_file):
    suppl = Chem.SDMolSupplier(sdf_file)
    mols = []
    smiles = []
    for mol in suppl:
        if mol is None:
            continue
        try:
            Chem.SanitizeMol(mol)
            smi = Chem.MolToSmiles(mol)
            mols.append(mol)
            smiles.append(smi)
        except:
            continue
    return mols, smiles

mols, smiles = load_fragments_from_sdf(sdf_path)

print(f"Loaded {len(mols)} valid fragments")
```

```python
# Display first 5 molecules from the fragment library
Draw.MolsToGridImage(mols[:5], molsPerRow=5, subImgSize=(200, 200))
```
